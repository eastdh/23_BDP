```python
from bs4 import BeautifulSoup
import urllib.request
```

```python
import re
result = []
sido1, sido2 = 0, 0
for sido1 in range(1, 18): # sido1의 변경
  for sido2 in range(1, 45): # sido2의 변경
      kyochon_url = f'https://www.kyochon.com/shop/domestic.asp?sido1={sido1}&sido2={sido2}&txtsearch=' 
      print(kyochon_url) 
      try:
        html = urllib.request.urlopen(kyochon_url) 
        soupKyochon = BeautifulSoup(html, 'html.parser') 
        tag_shopSchList = soupKyochon.find('div', 'shopSchList') 
      except:
        break
      #print(tag_shopSchList.prettify())
      for store in tag_shopSchList.find_all('span', 'store_item'):
        if len(store) <= 3:
          break 
        store_name = store.find('strong').string 
        store_em = re.split('[\n\t\r]', store.find('em').text)
        store_em = [w for w in store_em if w != '']
        store_fullAddr = store_em[0].split()
        store_sido = store_fullAddr[0]
        store_gungu = store_fullAddr[1]
        store_addr = ' '.join(store_fullAddr[2:])
        store_phone = store_em[-1]
        # print(store_name, store_sido, store_gungu, store_addr, store_phone)
        result.append([store_name]+[store_sido]+[store_gungu]+[store_addr]+[store_phone])
```

```python
print(len(result))
print(result[0])
print(result[-1])
```
